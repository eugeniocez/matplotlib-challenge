```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouseandstudy_df = pd.merge(study_results,mouse_metadata, on = 'Mouse ID')

# Display the data table for preview
mouseandstudy_df
```

```python
# Checking the number of mice.
micecount = mouseandstudy_df['Mouse ID'].nunique()
micecount
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse_ids = mouseandstudy_df[mouseandstudy_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]['Mouse ID'].unique()
duplicate_mouse_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data = mouseandstudy_df[mouseandstudy_df['Mouse ID'].isin(duplicate_mouse_ids)]
duplicate_mouse_data.sort_values(by=['Mouse ID', 'Timepoint'])
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_mouseandstudy_df = mouseandstudy_df[mouseandstudy_df['Mouse ID'].isin(duplicate_mouse_ids) == False]
cleaned_mouseandstudy_df
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_micecount = cleaned_mouseandstudy_df['Mouse ID'].nunique()
cleaned_micecount
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_stats = cleaned_mouseandstudy_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_stats = summary_stats.rename(columns={'mean':'Mean Tumor Volume',
                                              'median':'Median Tumor Volume',
                                              'var':'Tumor Volume Variance',
                                              'std':'Tumor Volume Std. Dev.',
                                              'sem':'Tumor Volume SEM'})
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_stats_advanced = cleaned_mouseandstudy_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(mean='mean',median='median',variance='var',std='std',sem='sem')

summary_stats_advanced
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
drug_regimen_count_pandas = cleaned_mouseandstudy_df['Drug Regimen'].value_counts()
drug_regimen_count_pandas.plot(kind='bar', figsize=(10,6),color='blue', alpha=0.7, align='center')

plt.xlabel('Drug Regimen')
plt.ylabel('# of Observed Mouse Timepoints')
plt.xticks(rotation=90)

plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
drug_regimen_count_pyplot = cleaned_mouseandstudy_df['Drug Regimen'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(drug_regimen_count_pyplot.index, drug_regimen_count_pyplot.values, color='b', alpha=0.7, align='center')

plt.xlabel('Drug Regimen')
plt.ylabel('# of Observed Mouse Timepoints')
plt.xticks(rotation=90)

plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
gender_count_pandas = cleaned_mouseandstudy_df['Sex'].value_counts()

gender_count_pandas.plot(kind='pie', figsize=(6,4), autopct='%1.1f%%', startangle=0, colors=['blue', 'orange'])

plt.axis('equal')
plt.ylabel('Sex')
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
gender_count_pyplot = cleaned_mouseandstudy_df['Sex'].value_counts()

plt.figure(figsize=(6,4))
plt.pie(gender_count_pyplot, labels=gender_count_pyplot.index, autopct='%1.1f%%', startangle=0, colors=['blue', 'orange'])

plt.axis('equal')
plt.ylabel('Sex')
plt.show()
```
